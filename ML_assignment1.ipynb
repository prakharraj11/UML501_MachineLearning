(f) For the matrix: n_array = np.array([[55, 25, 15],[30, 44, 2],[11, 45, 77]]), find
i. Sum of diagonal elements
ii. Eigen values of matrix
iii. Eigen vectors of matrix
iv. Inverse of matrix
v. Determinant of matrix

```python
n_array = np.array([[55,25,15],[30,44,2],[11,45,77]])
sum_diag = np.trace(n_array)
print(f"The sum of diagonal elements = {sum_diag}")
eigenVals, eigenVectors = np.linalg.eig(n_array)
print(f"Eigen values:\n{eigenVals}\nEigen Vectors are:\n{eigenVectors}")
inverted = np.linalg.inv(n_array)
print(f"Inverse:\n{inverted}")
det = np.linalg.det(n_array)
print(f"determinant of the matrix is given by = {det}")
```

# **Q2: Based on NumPy Mathematics and Statistics**

(a) For the array: array = np.array([[1, -2, 3],[-4, 5, -6]])
i. Find element-wise absolute value
ii. Find the 25th, 50th, and 75th percentile of flattened array, for each column, for each
row.
iii. Mean, Median and Standard Deviation of flattened array, of each column, and
each row

```python
array = np.array([[1, -2, 3], [-4, 5, -6]])
abs_array = np.abs(array)
print("Absolute array:")
print(abs_array)

# Flattened array
flat = abs_array.flatten()

p25_flat = np.quantile(flat, 0.25)
p50_flat = np.quantile(flat, 0.50)
p75_flat = np.quantile(flat, 0.75)
print(f"\nFlattened array percentiles: 25th = {p25_flat}, 50th = {p50_flat}, 75th = {p75_flat}")

p25_col = np.quantile(abs_array, 0.25, axis=0)
p50_col = np.quantile(abs_array, 0.50, axis=0)
p75_col = np.quantile(abs_array, 0.75, axis=0)
print(f"\nColumn-wise percentiles:")
print(f"25th: {p25_col}")
print(f"50th: {p50_col}")
print(f"75th: {p75_col}")

p25_row = np.quantile(abs_array, 0.25, axis=1)
p50_row = np.quantile(abs_array, 0.50, axis=1)
p75_row = np.quantile(abs_array, 0.75, axis=1)
print(f"\nRow-wise percentiles:")
print(f"25th: {p25_row}")
print(f"50th: {p50_row}")
print(f"75th: {p75_row}")

mean_flat = np.mean(flat)
median_flat = np.median(flat)
std_flat = np.std(flat)
print(f"\nFlattened array stats:\nMean = {mean_flat}, Median = {median_flat}, Std Dev = {std_flat}")

mean_col = np.mean(abs_array, axis=0)
median_col = np.median(abs_array, axis=0)
std_col = np.std(abs_array, axis=0)
print(f"\nColumn-wise stats:")
print(f"Mean: {mean_col}")
print(f"Median: {median_col}")
print(f"Std Dev: {std_col}")

mean_row = np.mean(abs_array, axis=1)
median_row = np.median(abs_array, axis=1)
std_row = np.std(abs_array, axis=1)
print(f"\nRow-wise stats:")
print(f"Mean: {mean_row}")
print(f"Median: {median_row}")
print(f"Std Dev: {std_row}")
```

```python
a = np.array([-1.8, -1.6, -0.5, 0.5, 1.6, 1.8, 3.0])

floor_values = np.floor(a)
ceil_values = np.ceil(a)
trunc_values = np.trunc(a)
round_values = np.round(a)

print("Original array:")
print(a)
print("\nFloor values:")
print(floor_values)
print("\nCeiling values:")
print(ceil_values)
print("\nTruncated values:")
print(trunc_values)
print("\nRounded values:")
print(round_values)
```

# **Q3: Based on Searching and Sorting**

(a) For the array: array = np.array([10, 52, 62, 16, 16, 54, 453]), find

i. Sorted array

ii. Indices of sorted array

iii. 4 smallest elements

iv. 5 largest elements

```python
array = np.array([10, 52, 62, 16, 16, 54, 453])

sorted_array = np.sort(array)
print("Sorted array:")
print(sorted_array)

sorted_indices = np.argsort(array)
print("\nIndices of sorted array:")
print(sorted_indices)

smallest_4 = np.partition(array, 4)[:4]
print("\n4 smallest elements:")
print(smallest_4)

largest_5 = np.partition(array, -5)[-5:]
largest_5_sorted = np.sort(largest_5)
print("\n5 largest elements:")
print(largest_5_sorted)
```

(b) For the array: array = np.array([1.0, 1.2, 2.2, 2.0, 3.0, 2.0]), find

i. Integer elements only

ii. Float elements only

```python
array = np.array([1.0, 1.2, 2.2, 2.0, 3.0, 2.0])

integer_elements = array[array == np.floor(array)]
print("Integer elements only:")
print(integer_elements)

float_elements = array[array != np.floor(array)]
print("\nFloat elements only:")
print(float_elements)
```

Q4:
(a) Write a function named img_to_array(path) that reads an image from a specified path and
save it as text file on local machine? (Note: use separate cases for RGB and Grey Scale
images)

```python
import cv2
import os

def img_to_array(path):
  img = cv2.imread(path, cv2.IMREAD_UNCHANGED)
  if img is None:
    print('Image not found')
    return
  base = os.path.splitext(os.path.basename(path))[0]

  if len(img.shape) ==2:
    print("Image is Greyscale")
    output_path = f'{base}_greyscale.txt'
    np.savetxt(output_path, img, fmt='%d')
    print(f"Grayscale image array saved to {output_path}")
  elif len(img.shape )==3 and img.shape[2]==3:
    print("Image is RGB")
    output_path = f"{base}_rgb.txt"
    h, w, _ = img.shape
    reshaped = img.reshape(-1, 3)
    np.savetxt(output_path, reshaped, fmt="%d")
    print(f"RGB image array saved to {output_path}")
  else:
    print("Unsupported image format.")

img_to_array('/content/IMG (1 of 2).jpg')
```
